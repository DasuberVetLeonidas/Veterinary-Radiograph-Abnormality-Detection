```python
import pandas as pd
import os
import subprocess
import numpy as np
```

```python
meta = pd.read_csv('../../data/AnimalDatasetV2Original/Meta For Animal Dataset Ver.2.csv',dtype={'Image No.':'str'})
```

```python
meta.drop(['Unnamed: 9'],axis=1,inplace=True)
```

```python
meta.head()
```

```python
meta['Normal_Abnormal'] = meta['Image Findings'] == 'Normal'
```

```python
dataset = meta.drop(['Patient No.','Species','Breed',
           'Age (yrs)','Gender','Neuter',
           'Image Findings','Thoracic or Pelvic'],
          axis=1
         )
```

```python
dataset['cv']=0
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset['Image No.'],
                                                    dataset['Normal_Abnormal'],
                                                    test_size = 0.2,
                                                    random_state = 42,
                                                    stratify = dataset['Normal_Abnormal']
                                                   )
print(len(X_test))
print(len(X_train))
for i in X_test:
    dataset.loc[dataset['Image No.']==i,'cv'] = 1
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train,
                                                    y_train,
                                                    test_size = 100/len(X_train),
                                                    random_state = 42,
                                                    stratify = y_train
                                                   )
print(len(X_test))
print(len(X_train))
for i in X_test:
    dataset.loc[dataset['Image No.']==i,'cv'] = 2
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train,
                                                    y_train,
                                                    test_size = 100/len(X_train),
                                                    random_state = 42,
                                                    stratify = y_train
                                                   )
print(len(X_test))
print(len(X_train))
for i in X_test:
    dataset.loc[dataset['Image No.']==i,'cv'] = 3
    
```

```python
def run_bash(command):
    process = subprocess.Popen(command.split(),stdout=subprocess.PIPE)
    output,error = process.communicate()
    return output,error
```

```python
dataset.head(n=20)
```

```python
normals = 0
abnormals = 0
for i,r in dataset.iterrows():
    if r['Normal_Abnormal']:
        normals += 1
        command = 'cp ' + '../../data/AnimalDatasetV2Original/' + r['Image No.'] + '.jpg ' + '../../data/AnimalDatasetV2CVs/CV' + str(r['cv']) + '/test/normal/'
        output,error = run_bash(command)
    else:
        abnormals += 1
        command = 'cp ' + '../../data/AnimalDatasetV2Original/' + r['Image No.'] + '.jpg ' + '../../data/AnimalDatasetV2CVs/CV' + str(r['cv']) + '/test/abnormal/'
        output,error = run_bash(command)
```

```python
for cv in np.arange(1,5+1):
    tmp_set = dataset[dataset['cv']!=cv]
    print(len(tmp_set))
    for i,r in tmp_set.iterrows():
        if r['Normal_Abnormal']:
            command = 'cp ' + '../../data/AnimalDatasetV2Original/' + r['Image No.'] + '.jpg ' + '../../data/AnimalDatasetV2CVs/CV' + str(cv) + '/train/normal/'
            output,error = run_bash(command)
        else:
            command = 'cp ' + '../../data/AnimalDatasetV2Original/' + r['Image No.'] + '.jpg ' + '../../data/AnimalDatasetV2CVs/CV' + str(cv) + '/train/abnormal/'
            output,error = run_bash(command)
```
